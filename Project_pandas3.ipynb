<h3> Replace salary <= 50, >=50 with binary

```python
df['income'].value_counts()
```

```python
sns.countplot(x='income',data=df)
```

```python
def salary_data(sal):
  if sal == '<=50':
    return 0
  else:
    return 1
```

```python
df['Enc_salary'] = df['income'].apply(salary_data)
```

```python
df
```

```python
df.replace(to_replace=['<=50K','>50K'],value=[0,1],inplace=True)
```
